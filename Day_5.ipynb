```python
arr3 = np.random.randint(0,10,(5,5))
print(arr3)
mean_4_0= np.mean(arr3, axis = 0) #column wise, from upside down
mean_4_1 = np.mean(arr3, axis = 1) #row wise, from left to right
```

```python
dice_1 = np.random.randint(1,7,size = 1000)
dice_2 = np.random.randint(1,7,size = 1000)
total = dice_1 + dice_2
frequency = np.bincount(total)
for i in range (2,13) :
    print(f"Sum {i} is appeared {frequency[i]} times")
```

```python
arr7 = np.random.randint(2,14, size = 15)
arr7_std = np.std(arr7)
arr7_mean = np.mean(arr7)
if arr7_std < (arr7_mean / 10) :
    print("the numbers are very close to each other")
if arr7_std > (arr7_mean * 0.3) :
    print("the numbers are spread from each other")    
        
```

```python
arr8 = np.random.randint(0,100, size = 20)
arr8_mean = np.mean(arr8)
arr8_std = np.std(arr8)
#if num is outlier abs(num - mean) > 2 * std, [usual formula]
#if we want to be stricker abs(num -mean) > 3 * std, [strong outlier]
mask = abs(arr8 -  mean) > 2 * arr8_std
arr8[mask] = arr8_mean
```
